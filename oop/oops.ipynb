```python
class Animal:
    def __init__(self,name_animal,lives_in):
        self.name = name_animal
        self.live = lives_in
        print('adding new Animals in the database:')

s1 = Animal("Dog","Land")
print(s1.name,s1.live)
s2 = Animal("Cat","land")
print(s2.name)
print(s2.live)
s3 = Animal('Dolphin','water')
print(s3.name)
print(s3.live)
```

```python
s2 = Animal()
print(s2.name)
```
